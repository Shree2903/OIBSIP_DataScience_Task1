```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("Iris.csv")
```

```python
df.head()
```

#### df.iloc[:,1:5] takes all rows and col 1 to 4
#### iloc selects rows and columns using positions 
#### df.iloc[:,5] takes species column only and all the rows

```python
x = df.iloc[:,1:5]
y = df.iloc[:,5]
```

#### Why Split Data?

#### We need:

#### some data for learning
#### some data for testing

#### Usually:

#### 80% training
#### 20% testing
#### What is random_state?

#### Used so results stay same every time.

```python
x_train,x_test,y_train,y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state = 42
)
```

#### What is KNN?
#### KNN = K Nearest Neighbors

#### It works like:

#### “Find nearest similar flowers and predict based on them.”

#### What is n_neighbors=3?

#### Means:

#### look at nearest 3 flowers
#### majority vote decides species

#### Example:

#### Neighbor	Species
#### 1	Setosa
#### 2	Setosa
#### 3	Virginica

#### Prediction:Setosa

```python
model = KNeighborsClassifier(n_neighbors = 3)
```

#### What Happens Here?

#### The computer studies patterns from training data.

#### This is called:Training.

```python
model.fit(x_train,y_train)
```

#### Make Predictions
#### What Happens Here?

#### The model predicts species for unseen flowers.

```python
predictions = model.predict(x_test)
```

#### Check Accuracy
#### Accuracy tells:

#### How many predictions were correct.

#### Example:

#### Accuracy: 1.0

#### means:

#### 100% correct

```python
accuracy = accuracy_score(y_test,predictions)
print("Accuracy:",accuracy)
```
